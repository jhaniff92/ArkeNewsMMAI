```python

#define a function to clean up the tweets. input - text field of all #the rows, output - cleaned text - remove @, hashtags, urls, emojis 

import re 

def cleanUpTweet(txt):
    # Remove mentions
    txt = re.sub(r'@+', '', txt)
    # Remove hashtags
    txt = re.sub(r'#[A-Z0-9]+', '', txt)
    # Remove retweets:
    txt = re.sub(r'RT : ', '', txt)
    # Remove urls
    txt = re.sub(r'https?:\/\/[A-Za-z0-9\.\/]+', '', txt)
    #remove amp
    txt = re.sub(r'&amp;', '', txt)
    #rempve strange characters
    txt = re.sub(r'Ã°Å¸â„¢', '', txt)
    #remove new lines
    txt = re.sub(r'\n', ' ', txt)
    return txt

df_final['Cleaned_Tweet'] = df_final['Tweet'].apply(cleanUpTweet)
```

## **Model**

### Train / Validation Split

```python
# Train Val Split - 75% Train, 25% Val
from sklearn.model_selection import train_test_split

X = df_final['Cleaned_Tweet']
y = df_final['Label_Number']


X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=42)
```

```python
print(X_train.shape)
print(X_val.shape)
print(y_train.shape)
print(y_val.shape)
```

### Part 1.a: Preprocessing and FE Pipeline

```python
#Count Vectorization 

from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer()
vect.fit(X_train)
```

```python
# learn training data vocabulary, then use it to create a document-term matrix
X_train_dtm = vect.transform(X_train)
```

```python
# combine fit and transform into a single step
X_train_dtm = vect.fit_transform(X_train)
```

```python
# examine the document-term matrix
X_train_dtm
```

```python
# transform testing data (using fitted vocabulary) into a document-term matrix
X_val_dtm = vect.transform(X_val)
X_val_dtm
```

```python
#TFID 

from sklearn.feature_extraction.text import TfidfTransformer

tfidf_transformer = TfidfTransformer()
tfidf_transformer.fit(X_train_dtm)
tfidf_transformer.transform(X_train_dtm)
```

### Part 1.b: Model Training/Tuning/Cross Validation 

```python
#Multinomial Naive Bayes

from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
```

```python
%time nb.fit(X_train_dtm, y_train)
```

```python
#Prediction

y_pred_class = nb.predict(X_val_dtm)

from sklearn import metrics
metrics.accuracy_score(y_val, y_pred_class)
```

```python
metrics.confusion_matrix(y_val, y_pred_class)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline

pipe = Pipeline([('bow', CountVectorizer()), 
                 ('tfid', TfidfTransformer()),  
                 ('model', MultinomialNB())])
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_val)
```

```python
metrics.accuracy_score(y_val, y_pred)
```
